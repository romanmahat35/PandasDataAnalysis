```python

import pandas as pd
df = pd.read_csv('/kaggle/input/stock-market-analysis-data/stocks.csv')
df.tail(10)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8),dpi=700, facecolor= "lightyellow")
plt.plot(df['Date'][:62],df['Low'][:62],color= "cornflowerblue")
plt.plot(df['Date'][:62],df['High'][:62],color= "blue")
plt.plot(df['Date'][62:124],df['Low'][62:124],color= "plum")
plt.plot(df['Date'][62:124],df['High'][62:124],color= "darkmagenta")
plt.plot(df['Date'][124:186],df['Low'][124:186],color= "paleturquoise")
plt.plot(df['Date'][124:186],df['High'][124:186],color= "cyan")
plt.plot(df['Date'][186:248],df['Low'][186:248],color= "lightcoral")
plt.plot(df['Date'][186:248],df['High'][186:248],color= "darkred")

plt.xlabel("Date",fontdict= {'family':'fantasy','color':'black','size':15})
plt.xticks(rotation=90, ha='right')
plt.ylabel("Stock Price",fontdict= {'family':'fantasy','color':'black','size':15})
plt.title("COMPARISION OF LOW & HIGH STOCK PRICES OF 4 COMPANIES", fontdict= {'family':'fantasy','color':'black','size':20})

plt.legend(['Low Apple','High Apple' ,'Low Microsoft','High Microsoft' ,'Low Netflix','High Netflix','Low Google','High Google'], loc ="center right")
plt.show()
```

```python
avg_trade ={
    'Company' : ['Apple' ,'Microsoft','Netflix','Google'],
    'Avg Volume' : [round(df['Volume'][:62].mean()), round(df['Volume'][62:124].mean()),round(df['Volume'][124:186].mean()),round(df['Volume'][186:248].mean())]
            }
df_avg_trade = pd.DataFrame(avg_trade)
df_avg_trade
```

```python
fig = plt.figure(figsize = (8, 5))
barplot = plt.bar(df_avg_trade['Company'], df_avg_trade['Avg Volume'],color =['red', 'yellow', 'pink', 'cyan'] ,edgecolor = "black", width = 0.25)
plt.bar_label(barplot , labels= df_avg_trade['Avg Volume'], label_type='edge')
plt.xlabel("Company",fontdict= {'family':'fantasy','color':'black','size':10})
plt.ylabel("Volume",fontdict= {'family':'fantasy','color':'black','size':10})
plt.title("COMPANY & AVERAGE TRADE VOLUME",fontdict= {'family':'fantasy','color':'black','size':15})
plt.show()
```
